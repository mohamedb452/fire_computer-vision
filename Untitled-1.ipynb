```python
from ultralytics import YOLO 
import cv2
model=YOLO("yolov8s.pt")


results=model.train(
    data="data.yaml",
    epochs=15,
    imgsz=416,
    batch=25,
    name="fire_dector",
    optimizer="Adam",
    patience=4,
)
```

```python

from ultralytics import YOLO
import cv2

# تحميل النموذج المدرب
model = YOLO("my_fire_model_.pt")
# فتح الفيديو
video_path = "E:\Artificial intelligence\computer vision\project\istockphoto-1490073588-640_adpp_is.mp4"
cap = cv2.VideoCapture(video_path)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    #تشغيل الكشف باستخدام النموذج
    results = model(frame)
    # رسم النتائج على الإطار
    annotated_frame = results[0].plot()
    # عرض الإطار
    cv2.imshow("Fire Detection", annotated_frame)
    # اضغط 'q' للخروج
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# تحرير الموارد
cap.release()
cv2.destroyAllWindows()
```
